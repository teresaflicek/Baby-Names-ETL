## Cleaning up second dataset: Top 25 California Baby Names in 2018

```python
#top 25 baby names in California

#read csv file
file = "resources/top-25-baby-names-by-gender-2005-current.csv"
top_names = pd.read_csv(file)

#display dataframe
top_names.head()
```

```python
#display just the top 25 names in CA for the year 2018
top_names_CA = top_names[(top_names.Year == 2018)]

#capitalize the Name column values and create copy
top_names_CA['Name'] = top_names_CA['Name'].str.capitalize().copy()

#display dataframe
top_names_CA
```

```python
#rename the columns to differentiate from the national dataset and create a copy
top_names_CA = top_names_CA.rename(columns={"Rank": "CA Rank", "Year": "CA Year", "Gender": "CA Gender", "Name": "CA Name", "COUNT": "CA Count"}).copy()

#display dataframe
top_names_CA
```

```python
#sort the CA Rank column by values so the top female and male names are displayed next to each other and create a copy to continue using same df variable
top_names_CA = top_names_CA.sort_values(by=['CA Rank']).copy()

#display dataframe
top_names_CA
```
